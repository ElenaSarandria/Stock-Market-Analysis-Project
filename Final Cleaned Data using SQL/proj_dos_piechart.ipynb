```python
# Import our dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
stocks = pd.read_csv('project_dos.csv')
stocks.head()
```

```python
stocks9 = stocks[stocks['pct_change'] >0]
stocks9
```

```python
stocks10 = stocks.sort_values(['pct_change'], ascending=True)
stock10h = stocks10.head(10)
stock10h
```

```python
ticker = stocks9['symbol'].unique()
ticker
```

```python
ticker10 = stock10h['symbol'].unique()
ticker10
```

```python
pct_chg = stocks9['pct_change'].unique()
pct_chg
```

```python
pct_chg10 = stock10h['pct_change'].unique()
pct_chg10
```

```python
plt.pie(pct_chg, labels=None,pctdistance=1.10, autopct="%1.1f%%", shadow=True, startangle= 0)
centre_circle = plt.Circle((0,0),0.75,fc='#e0e0e0')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal') 
plt.tight_layout()
plt.legend(ticker,
           title="Positive Stocks during Crisis",
          loc="center left",
          bbox_to_anchor=(1, 0, .5, 1))
plt.show()
```

```python
plt.pie(pct_chg10, labels=None,pctdistance=1.1, autopct="%1.1f%%", shadow=True, startangle= 0)
centre_circle10 = plt.Circle((0,0),0.75,fc='#e0e0e0')
fig2 = plt.gcf()
fig2.gca().add_artist(centre_circle10)
plt.axis('equal') 
plt.tight_layout()
plt.legend(ticker10,
           title="Bottom 10 Stocks during Crisis",
          loc="center left",
          bbox_to_anchor=(1, 0, .5, 1))
plt.show()
```
