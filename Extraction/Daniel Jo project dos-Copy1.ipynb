```python
# symbol_list = df_symbol.tolist()
# print (symbol_list)

symbol_list = df["Symbol"].tolist()
print(symbol_list)
```

```python
# /v2/everything

ag_results = []
for symbols in symbol_list:
    

    every_articles = newsapi.get_everything(q= (symbols),
                                          from_param='2020-02-09',
                                          to='2020-03-08',
                                          language='en',
                                          sort_by='relevancy')
    results = {
        "symbol": symbols,
        "article_count": every_articles['totalResults']
    }
    ag_results.append(results)
new_data = pd.DataFrame(ag_results)

new_data.head()
    
# every_articles['totalResults']
```

```python
ag_stock = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close","Volume", "Divendends", "Stock Splits",\
                                "Symbol"])
symbols
```

```python
for symbols in symbol_list:
    results = yf.Ticker(symbols)
    stock_data = results.history(period="max")
    symbol_df = pd.DataFrame(stock_data)
    symbol_df['Symbol'] = symbols
    symbol_df['Date'] = symbol_df.index
    # drop index
    symbol_df.reset_index(drop=True)
    # append with out index 
    ag_stock = ag_stock.append(symbol_df,ignore_index=True)


```
